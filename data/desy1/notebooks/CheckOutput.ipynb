```python
import numpy as np
import pylab as mplot
import sacc
%pylab inline
```

```python
dir = '../outputs/'

file = 'twopoint_data_real.sacc' 
d = sacc.Sacc.load_fits(dir+file)

file = 'twopoint_theory_real.sacc'
t = sacc.Sacc.load_fits(dir+file)

W = "galaxy_density_xi"
GAMMA = "galaxy_shearDensity_xi_t"
GAMMAX = "galaxy_shearDensity_xi_x"
XIP = "galaxy_shear_xi_plus"
XIM = "galaxy_shear_xi_minus"
EE = "galaxy_shear_cl_ee"
DD = "galaxy_density_cl"
ED = "galaxy_shearDensity_cl_e"
```

```python
mplot.figure(figsize=(8,10))

for i in range(4):
    mplot.subplot(4,2,i*2+1)
    theta, xi, cov = d.get_theta_xi(XIP, 'source_'+str(i), 'source_'+str(i), return_cov = True)
    rr = np.sqrt(np.diag(cov))
    mplot.errorbar(theta, theta*xi, theta*rr)
    mplot.xscale('log')
    # mplot.yscale('log')
    mplot.ylim(-1e-4, 6e-4)
    
    mplot.subplot(4,2,i*2+2)
    theta, xi, cov = d.get_theta_xi(XIM, 'source_'+str(i), 'source_'+str(i), return_cov = True)
    rr = np.sqrt(np.diag(cov))
    mplot.errorbar(theta, theta*xi, theta*rr)

    mplot.xscale('log')
    # mplot.yscale('log')
    mplot.ylim(-1e-4, 6e-4)

mplot.tight_layout()
```

```python
mplot.figure(figsize=(4,10))

for i in range(5):
    mplot.subplot(5,1,i+1)
    theta, xi, cov = d.get_theta_xi(W, 'lens_'+str(i), 'lens_'+str(i), return_cov = True)
    rr = np.sqrt(np.diag(cov))
    mplot.errorbar(theta, theta*xi, theta*rr)
    mplot.xscale('log')
    # mplot.yscale('log')

mplot.tight_layout()
```
