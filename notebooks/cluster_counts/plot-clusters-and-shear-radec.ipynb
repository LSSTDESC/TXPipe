```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import os
```

```python
# Get the path to the data files
data_dir = "data/cosmodc2/20deg2"

#Â open the cluster and shear catalogs
clusters = h5py.File(os.path.join(data_dir, "cluster_catalog.hdf5"), "r")
shear = h5py.File(os.path.join(data_dir, "shear_catalog.hdf5"), "r")
```

```python
# Get the RAs and Decs of the clusters and galaxies
cluster_ra = clusters['clusters/ra'][:]
cluster_dec = clusters['clusters/dec'][:]
gal_ra = shear["shear/00/ra"][:]
gal_dec = shear["shear/00/dec"][:]
```

```python
# Plot the galaxies and clusters
#Â Note the overspilling source galaxies!
plt.plot(gal_ra, gal_dec, ',', label="Galaxies")
plt.plot(cluster_ra, cluster_dec, 'x', label="Clusters")
```
