```python
import sys
import os
import ceci
#Â get current directory
import os
notebook_dir = os.path.abspath("")
txpipe_dir = os.path.join(notebook_dir, "../..")
sys.path.append(txpipe_dir)
import txpipe
```

```python
# These six pixels come to about 23 deg^2
pixels = [8786, 8787, 8788, 8914, 8915, 9043]
outdir = "./data/cosmodc2/inputs-20deg2"
extra_cols="redshift_true"
response_model="{txpipe_dir}/data/DESY1-R-model.hdf5"

# prepare to run
os.makedirs(outdir, exist_ok=True)

# configure the stage
stage = txpipe.TXCosmoDC2Mock.make_stage(
    response_model=response_model,
    shear_catalog=f"{outdir}/shear_catalog.hdf5",
    photometry_catalog=f"{outdir}/photometry_catalog.hdf5",
    healpixels=pixels,
    extra_cols=extra_cols
)
```

```python
# run the stage
stage.run()

# move the stage outputs to their final locations
stage.finalize()
```
